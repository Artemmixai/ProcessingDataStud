# Построение простейших графиков, знакомство с библиотекой matplotlib

Библиотека `matplotlib` -- это мощный графический инструмент языка `Python`, позволяющий визуализировать научные данные в виде разнообразных графиков (от простых `y(x)` до детализированных контурных карт) и подготавливать качественные рисунки, иллюстрации и даже анимированные изображения. Библиотека богата различными инструментами и cостоит из множества модулей, каждый со своими классами и функциями.

Для желающих прочувствовать полный функционал, можем предложить [онлайн-учебник Павла Шабанова](https://github.com/whitehorn/Scientific_graphics_in_python). Здесь рассмотрим "краткое содержание", необходимое для наших задач

Импортируем `matplotlib.pyplot` с общепринятым сокращением `plt` и `Numpy` для создания данных

```python
import matplotlib.pyplot as plt
import numpy as np
```

Начнём с простейшего построения двумерного графика. Для него сгенерируем некоторые данные и отобразим связкой функций `plt.plot()` - создание графика и `plt.show()` - отображение окна

```python
x = np.linspace(0, 3., 10)
y = x**2. 

plt.plot(x, y)
plt.show()
```

В **Jupyter Notebook** последнюю строку можно опускать. Вся суть именно в первой функции, принимющая огромное количество аргументов, с которыми будем знакомиться по ходу модуля. Но и это ещё не всё - сгенерированный выше график был создан всего одной командой, однако, как уже упоминалось выше, на самом деле "за кулисами" было произведено сразу несколько действий:

1. было создано полотно, или "канва" (FigureCanvas), -- интерфейс, который задает область будущего изображения;
2. к созданной канве был прикреплен высокоуровневый объект "фигура" (Figure Artist), который будет вмещать в себя все элементы будущего графика;
3. на созданную фигуру был наложен объект "оси" (Axes Artist), на котором определено большинство методов библиотеки `matplotlib`;
4. и, наконец, на осях был изображен график (который, в свою очередь, состоит из примитивных элементов, таких как линии, точки, окружности, прямоугольники, текст).

Все эти действия за нас выполнил скрипт, реализованный в разделе `matplotlib.pyplot`. Отметим также, что диапазон и масштаб по обеим осям координат метод подобрал автоматически по переданным ему данным -- при изменении диапазона данных график автоматически перестроится.

### plt.subplot(rows, columns, index)

Поговорим о возможности отрисовки нескольких графиков в одном окне. Для этого воспользуемся `plt.subplot(rows, columns, index)`. Первый аргумент задаёт число строк, на которое разбивается окно, второй - число солбцов. А третий определяет положение текущего элемента

Если передать число *N*, то в квадрате с координатами *((N-1)%columns, (N-1)//columns+1)*

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
plt.plot(x, y)

plt.subplot(2, 3, 2)
plt.subplot(2, 3, 6)
```

Также третьим аргументом может быть кортеж из двух чисел, указывающий на номера левого верхнего и правого нижнего квадратов: 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
plt.plot(x, y)

plt.subplot(3, 3, (1, 5))
```

Более того, ничего не мешает накладывать несколько сеток на один график

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
plt.plot(x, y)

plt.subplot(2, 3, 3)
plt.subplot(3, 4, 9)
```

Дополним заполнением всего пространства окнами одинакового размера, где координаты задаётся с помощью индексов(начало из [0, 0])

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
fig, axs = plt.subplots(2, 3)
axs[0, 0].plot(x, y)
axs[1, 2].plot(x, y)

plt.tight_layout()  # Улучшит видимость границ окон
```

### plt.figure(num, figsize: tuple, dpi)

Одновременное создание нескольких окон обеспечивается данной функцией. Сейчас выделим три аргумента:
- *num* - идентификатор окна
- *figsize: tuple* - размеры графического окна в дюймах
- *dpi: float* - плотность пикселей

```python
fig1 = plt.figure(1, figsize=(10, 10), dpi=30)
ax1 = fig1.add_subplot()
fig2 = plt.figure(2, figsize=(10, 10), dpi=200)
ax2 = fig2.add_subplot()

ax1.plot(x, y)
ax2.plot(-x, y)
```

Применение `.add_subplot(rows, columns, index)` отличается от `plt.subplot(rows, columns, index)` объектом применения. Первый требует фигуру, тогда как вторая добовляет окно в текущую фигуру

~~~python
ax1 = plt.subplot(2, 3, 1)
ax2 = plt.subplot(2, 3, 6)
~~~
или
~~~python
fig = plt.figure()
ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(2, 3, 6)
~~~
содержимое *ax1, ax2* попарно идентично

```python
ax1 = plt.subplot(2, 3, 1)
# Доступно сокращение в виде трёхзначного числа 
ax2 = plt.subplot(236)

fig = plt.figure()
ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(236)

# Добавят в последнее созданное окно
plt.subplot(2, 3, 2)
plt.subplot(2, 3, 3)
```

Двумерными линейными графиками функционал библиотеки не ограничивается. Быстро перейдём в полярные координаты функцией `plt.polar(angle, R, ...)`

#### Гистограммы

Парой нельзя обойтись без применения гистограмм. Гистограмма состоит из примыкающих прямоугольников, расположенных вдоль оси x, которые разбиты на дискретные интервалы, их называют *bins*. Их площадь пропорциональна частоте конкретного интервала. Такой способ визуализации часто используют в статистике для демонстрации распределения.Для них существует функция `plt.hist(massive, bins)`

```python
massive = np.random.randint(0, 100, 100)
amount, avg, patch = plt.hist(massive, bins=20, histtype='bar')
print(amount)
print(avg)
```

Результатом работы является не только сам график, но и рассчитанное количество чисел в каждом столбце и их среднее значение

#### Столбчатые диаграммы

Еще один распространенный тип графиков — столбчатые диаграммы. Они похожа на гистограммы, но на оси x тут располагаются не числовые значения, а категории. В matplotlib для реализации столбчатых диаграмм используется функция `bin()`

```python
index = np.arange(5)
values1 = [5,7,3,4,6]
plt.bar(index, values1)
```

Или в горизонтальном виде

```python
index = np.arange(5)
values1 = [5,7,3,4,6]
plt.barh(index, values1)
```

#### Контурные карты
*Контурной картой* функции f(x, y) называют совокупность линий уровня f(x, y) = const на плоскости. Контуры могут образовать причудливую конфигурацию вложенных/замкнутых/разомкнутых/... линий, относительное расположение которых позволяет изучать поведение функции в области её определения. В задачах электростатики линии уровня называются *эквипотенциалями*. В теории поверхностей линии получили название *геодезических*, поскольку на картах подобными линиями обозначается связанное множество точек на карте, расположенных на одной высоте.

Создадим контурную карту на основе двумерной сетки, которой в соответсвие рассчитаем значения некоторой функции от двух координат `plt.contour(X, Y, Z, level: int, ...)`. Последний аргумент задаёт число отображаемых линий уровня

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1., 100)
y = np.linspace(-1, 1., 100)
X, Y = np.meshgrid(x, y)  # функция преобразует одномерные массивы в двумерную сетку
func = X*Y

ptr = plt.contour(X, Y, func, levels=40, vmin=-1, vmax=0)
# добавим подписи значений функции для каждой линии уровня
plt.clabel(ptr, fontsize=10)

# полный список параметров, которые можно передать в функцию, можно найти по
# адресу https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.contour.html
```

#### Тепловые карты
Другим средством двумерного представления трёхмерных данных является использование тепловых карт.

*Тепловая карта* - график на плоскости, в котором для представления значения функции используется цвет, так что точки на плоскости, отвечающие условию f(x,y) = const, окрашены одинаковым образом.

Главное отличие тепловой карты от контурной - в непрерывности представления данных. Тепловая карта может быть получена из контурной закрашиванием пространства между двумя соседними линиями уровня.

Для отрисовки тепловой карты используется функция `contourf`. А вместо легенды, добавим цветовую шкалу `plt.colorbar` Синтаксис и пример приведены ниже.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1., 10)
y = np.linspace(-1, 1., 10)
X, Y = np.meshgrid(x, y)  # функция преобразует одномерные массивы в двумерную сетку
func = X*Y

ptr = plt.contourf(X, Y, func, levels=10, vmin=-1, vmax=1)
plt.colorbar(ptr, ticks=[i for i in np.linspace(-1., 1, 11)])

# список параметров, которые можно передать в функцию, почти аналогичен функции contour и может быть найден по адресу
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.contourf.html#matplotlib.pyplot.contourf
```

*rstride* и *cstride* задают шаг в изображении точек по строкам и столбцам соответственно

```python
import numpy as np
import matplotlib.pyplot as plt

L, n = 2, 400
x = np.linspace(-L, L, n)
y = x.copy()
X, Y = np.meshgrid(x, y)  # Создали двумерную сетку
Z = np.exp(-(X**2 + Y**2))

# В первых двух аргументах пропущены записи "nrows=" и "ncols="
fig, ax = plt.subplots(1, 2, subplot_kw={'projection': '3d'})

ax[0].plot_wireframe(X, Y, Z, rstride=40, cstride=40)
ax[1].plot_surface(X, Y, Z, rstride=40, cstride=40)
```

Наконец упомянем изменение стороны отображения на трёхмерных графиках. `ax.view_init()` задаёт направление обзора в градусах

```python
import numpy as np
import matplotlib.pyplot as plt

L, n = 2, 400
x = np.linspace(-L, L, n)
y = x.copy()
X, Y = np.meshgrid(x, y)  # Создали двумерную сетку
Z = np.exp(-(X**2 + Y**2))

fig, ax = plt.subplots(1, 2, subplot_kw={'projection': '3d'})
ax[0].plot_wireframe(X, Y, Z, rstride=40, cstride=40)
ax[0].view_init(35, 90)

ax[1].plot_wireframe(X, Y, Z, rstride=40, cstride=40)
ax[1].view_init(65, 65)
```
