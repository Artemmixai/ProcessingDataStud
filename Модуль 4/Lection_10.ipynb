## Линии и маркеры

Предлагаю сгенерировать один объём данных, с которым будем работать в течение занятия

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 10, 10)
print(x)
y_1 = np.sin(x)
print(y_1)
y_2 = np.cos(x)*2
print(y_2)
````

Начнём с простого двумерного графика, на котором отобразим обе линии. Речь пойдёт о функции `plt.plot`, но для `plt.scatter` всё аналогично

````python
plt.plot(x, y_1)
plt.plot(x, y_2)
````

Автоматически для каждой линии выбран свой цвет. При большем количестве может быть трудно различать оттенки, поэтому перейдём к другим типам самих линий (*linestyle*) и настроим цвет (*color*)

А еще в ```matplotlib``` встроено некоторое количество "названных" цветов (например, salmon, darkseagreen, indigo, и т. д.). Полный список можно найти [здесь](https://matplotlib.org/stable/gallery/color/named_colors.html).

Пусть первая линия будет пунктирной синего цвета, а вторая штрихпунктирной красного. Дополнительно изменим толщинц линий (*linewidth*). Для этого укажем следующие аргументы

````python
plt.plot(x, y_1, linestyle=':', color='blue', linewidth=3)
plt.plot(x, y_2, linestyle='-.', color='r', linewidth=4)
````

В аргументе *color* можно передавать первую букву цвета, название на английском, RGB кортеж или целую цветовую палитру, если одной функцией отстраиваются несколько линий

Видимая линия - попытка функции соединить все точки, а ведь именно они являются результатом. Выделим их некоторым маркером и настроим размер (*markersize*), цвет (*color*), обводку (*markeredgecolor* и *markerwdgewidth*). 

Для разделения, отрисуем маркеры отдельной функцией, хотя можно всё настраивать в одной

````python
plt.plot(x, y_1, linestyle=':', color='blue', linewidth=3)
plt.plot(x, y_1, 'o', markerfacecolor='r', markersize=7, markeredgecolor='k', markeredgewidth=2)

plt.plot(x, y_2, linestyle='-.', color='r', linewidth=4)
plt.plot(x, y_2, 'D', color='black', markersize=10, markeredgecolor='orange',  markeredgewidth=5)
````

Сегодня намеренно делаем вырвиглазное оформления для наблюдения эффекта от вносимых изменений - в жизни не делайте так)

Об остальных возможностях *plt.plot* можно почитать на [официальном сайте](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

## Легенда

Вернёмся к приемлемому виду и поговорим о легенде. Она позволяет отобразить значение каждой линии. Вызовем командой `plt.legend()` с добавлением метки к каждой линии (*label*)

````python
plt.plot(x, y_1, linestyle=':', color='blue', linewidth=3, label='sin')
plt.plot(x, y_2, linestyle='-.', color='r', linewidth=3, label='2*cos')
plt.legend()
````

Второй вариант задания легенды - перечислить названия кривых в массие и передать через `plt.legend`

````python
plt.plot(x, y_1, linestyle=':', color='blue', linewidth=3)
plt.plot(x, y_2, linestyle='-.', color='r', linewidth=3)
plt.legend(['sin', 'cos'])
````

Настроим её: зададим местоположение (*loc*) и размер шрифта (*fontsize*)

Допонительно: число колонок (*ncolsint*), цвет текста (*labelcolor*), поменяем порядок маркера и подписи (*markerfirst*), прозрачность (*framealpha*) и воспроизведём в обратной последовательности линии (*reverse*)

# Погрешности

Любые измерения или расчеты сопровождаются теми или иными ошибками, приводящими к возникновению погрешностей. При представлении результатов работы обязательно нужно указывать, с какой точностью были получены и рассчитаны данные. Это особенно важно, когда при обработке данных производится подгон некой функции к зависимости (фит), извлекается некая новая величина или оценивается, в какой точке график пересечет ту или иную ось. Для отображения погрешностей данных на графиках в `matplotlib` предусмотрен метод `errorbar`.

Продемонстрируем его сразу на примере.

````python
import numpy as np

x = np.linspace(1, 10, 40)
y_1 = np.sin(x)
plt.plot(x, y_1, linestyle=':', color='blue', linewidth=1, label='Initial')
# Добавим к каждой точке случайный "дребезг" -- в пределах 5% от максимального значения функции
y_1 += np.max(y_1) * (1 - 2*np.random.rand(len(y_1)))*0.05
# Создадим массив с погрешностями -- для примера зададим их случайно, в пределах 4% от максимума
yerr = np.random.rand(len(y_1))*np.max(y_1)*0.04

# Нарисуем погрешности для некоторых точек и обозначим эти точки красными пустыми кружками
# ## Добавлять маркеры вовсе не обязательно -- попробуйте убрать соответствующие параметры

# capsize -- величина "шапочек" погрешностей
# errorevery -- рисовать погрешность не для каждой точки, а для каждой n-ой
# linestyle='' -- отключает рисование линий на графике (уберите его и посмотрите, что будет)
# ecolor -- цвет обозначения погрешностей
# markevery -- добавлять маркеры не к каждой точке
# ms  -- то же самое, что markersize
# mfc -- то же самое, что markerfacecolor
# mec -- то же самое, что markeredgecolor
plt.errorbar(x, y_1, yerr, capsize=4, errorevery=1, linestyle='', ecolor='r',
             marker='', markevery=10, ms=5, mfc='none', mec='r', label='error')

plt.legend()
````

Обсудим общее оформление графиков. Первое, на что обращаем внимание, глядя на иззображание, это на название, содержащее информацию о том, на что мы вообще смотрим. Как не удивительно, заголовок задаётся методом `plt.title(title: str)`

````python
plt.title('Graph name', loc='center', fontsize=18)
y_1 = np.sin(x)
y_2 = np.cos(x)*2
plt.plot(x, y_1)
plt.plot(x, y_2)
````

Первым аргументом передали название, отображаемое в верхней части рисунка, 

Вторым - выравнивание (`loc="center" | "left" | "right"`), 

Третий отвечает за размер шрифта

Идентично подпишем оси:

````python
plt.title('Graph name', loc='center', fontsize=18)
plt.xlabel('X axis', loc='right', fontsize=15)
plt.ylabel('Y axis', loc='bottom', fontsize=15)

plt.plot(x, y_1)
plt.plot(x, y_2)
````

Отличие в местоположении вертикальной оси - доступны положения `"bottom", "center", "top"`

Ограничим область видимости, чтобы выделить фрагмент. `plt.xlim(left: float, right: float)` и `plt.ylim(left: float, right: float)` позволят сделать это

````python
plt.title('Graph name', loc='center', fontsize=18)
plt.xlabel('X axis', loc='right', fontsize=15)
plt.ylabel('Y axis', loc='bottom', fontsize=15)
plt.xlim(3, 7)
plt.ylim(-1, 1)

plt.plot(x, y_1)
plt.plot(x, y_2)
````

Есть возможность добавить надпись в любую часть рисунка. `plt.text(x, y, text, fonsize...)`

````python
plt.text(2, 1.5, 'Text', fontsize=18, color='orange')

plt.plot(x, y_1)
plt.plot(x, y_2)
````

Для упрощения определения значений отдельных точек создадим сетку. `plt.grid(which= "major" | "minor" | "both", axis = "both" | "x" | "y", color, linewidth...)`. Аргумент __which__ задаёт тип линий, подробнее будет понятнее позже. __axis__ настраивает от какой оси будут проведены линии

````python
plt.title('Graph name', loc='center', fontsize=18)
plt.xlabel('X axis', loc='right', fontsize=15)
plt.ylabel('Y axis', loc='bottom', fontsize=15)
plt.xlim(3, 7)
plt.ylim(-1, 1)
plt.grid(which='both', axis='both')

plt.plot(x, y_1)
plt.plot(x, y_2)
````

Плотнее займёмся осями. У нас есть возможность исправлять шрифт, размер тиков, отрисовываемые числовые значения. Пойдём по порядку:

Размер шрифта задаётся тем же аргументом _fontsize_

_ticks_ принимает значения, для которых будет отображатся число на оси

_labels_ задаёт подписи к ранее переданным значениям, т.е. вместо _0.5_ можно оставить _1/2_

````python
plt.title('Graph name', loc='center', fontsize=18)
plt.xlabel('X axis', loc='right', fontsize=15)
plt.ylabel('Y axis', loc='bottom', fontsize=15)
plt.xlim(3, 7)
plt.ylim(-1, 1)
plt.grid(which='both', axis='both')

plt.xticks(fontsize=12,
           ticks=[3, 4, 5, 6, 7])
plt.yticks(ticks=[1, 0.5, 0, -0.5, -1], 
           labels=['1', '1/2', '0', '-1/2', '-2/2'], 
           fontsize=18)

plt.plot(x, y_1)
plt.plot(x, y_2)
````

Дополнительно поработаем со штрихами рядом с подписями через `plt.tick_params(axis = "both" | "x" | "y", which= "major" | "minor" | "both", width, length)`. Выбираем ось для настройки и тип тиков аналогично `plt.grid()`, и добавляем их размер

````python
from matplotlib import ticker


ax = plt.figure().add_subplot()
plt.title('Graph name', loc='center', fontsize=18)
plt.xlabel('X axis', loc='right', fontsize=15)
plt.ylabel('Y axis', loc='bottom', fontsize=15)
plt.xlim(3, 7)
plt.ylim(-1, 1)
plt.grid(which='both', axis='both')

plt.xticks(fontsize=14,
           ticks=[3, 4, 5, 6, 7])
plt.yticks(ticks=[1, 0.5, 0, -0.5, -1], 
           labels=['1', '1/2', '0', '-1/2', '-2/2'], 
           fontsize=18)

plt.tick_params(axis="x", which='major', width=2.00, length=10)
plt.tick_params(axis="y", which='major', width=6.00, length=5)

plt.plot(x, y_1)
plt.plot(x, y_2)
````

Так всё таки, что значит *major*? Под ним мы понимаем тики, которым поставлено в соответствие число, отображаемое на рисунке. Добавим к ним промежуточные значения *minor*

````python
from matplotlib import ticker


ax = plt.figure().add_subplot()
plt.title('Graph name', loc='center', fontsize=18)
plt.xlabel('X axis', loc='right', fontsize=15)
plt.ylabel('Y axis', loc='bottom', fontsize=15)
plt.xlim(3, 7)
plt.ylim(-1, 1)
plt.grid(which='major', axis='both', linewidth=0.5, color='k')
plt.grid(which='minor', axis='both', linewidth=1, color='r')

plt.xticks(fontsize=12,
           ticks=[3, 4, 5, 6, 7])
plt.yticks(ticks=[1, 0.5, 0, -0.5, -1], 
           labels=['1', '1/2', '0', '-1/2', '-2/2'], 
           fontsize=18)

plt.tick_params(which='major', width=2.00, length=5)
plt.tick_params(which='minor', width=1.25, length=3.5)

ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())

plt.plot(x, y_1)
plt.plot(x, y_2)
````

Сразу же внёс изменения в сетку. `ax.xaxis.set_minor_locator` требует применения к созданному элементу окна, которое ввели. Принимает в качестве аргумента определённый класс, который задаёт количество и местоположение тиков. Обо всех их вариантах прочитайте [здесь](https://matplotlib.org/stable/users/explain/axes/axes_ticks.html)

Специальные символы удобнно включать через `latex`, о котором можно почитать в приложенном учебнике. Для этого добавим `plt.rc('text', usetex=True)`

````python
from matplotlib import ticker

plt.rc('text', usetex=True)

ax = plt.figure().add_subplot()
plt.title('Graph name', loc='center', fontsize=18)
plt.xlabel('X axis', loc='right', fontsize=15)
plt.ylabel('Y axis', loc='bottom', fontsize=15)
plt.xlim(3, 7)
plt.ylim(-1, 1)
plt.grid(which='major', axis='both', linewidth=0.5, color='k')
plt.grid(which='minor', axis='both', linewidth=1, color='r')

plt.xticks(fontsize=12,
           ticks=[3, 4, 5, 6, 7])
plt.yticks(ticks=[1, 0.5, 0, -0.5, -1], 
           labels=['1', r'$\frac{1}{2}$', '0', 
                   r'$-\frac{1}{2}$', r'$-\frac{2}{2}$'], 
           fontsize=18)

plt.tick_params(which='major', width=2.00, length=5)
plt.tick_params(which='minor', width=1.25, length=3.5)

ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())

plt.plot(x, y_1)
plt.plot(x, y_2)
````
